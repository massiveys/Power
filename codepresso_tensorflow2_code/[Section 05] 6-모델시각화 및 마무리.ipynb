```
import tensorflow as tf
from tensorflow import keras
```

## 모델 시각화
* Graph Network 객체인 모델은 `keras.utils.plot_model()` 함수를 이용하여 모델의 시각화 가능  
  * `Sequentail` API  
  * `Functional` API

```
# 모델 구성

inputs = keras.Input(shape=(28, 28, 1)) # Input을 선언

x = keras.layers.Flatten()(inputs)
x = keras.layers.Dense(256, activation='relu')(x)
x = keras.layers.Dense(128, activation='relu')(x)
x = keras.layers.Dropout(rate=0.2)(x)
predictions = keras.layers.Dense(10, activation='softmax')(x)
```

```
# 모델 객체 생성

model = keras.Model(
    inputs=inputs,
    outputs=predictions
)
```

```
keras.utils.plot_model(model, 'first_name.png')
```

```
keras.utils.plot_model(model, 'first_name.png', show_shapes=True)
```
